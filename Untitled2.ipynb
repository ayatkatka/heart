```python
import pandas as pd
df=pd.read_csv('heart.csv')
print(df)
print(df.info())
print(df.describe())
print(df.loc[0:15, ['Age','Cholesterol']])
print(df.head())
print(df.tail())
print(df.shape)


print(df[['Cholesterol','MaxHR']])
print(df[df['Cholesterol']>400])
```

```python
#data visualization
import matplotlib.pyplot as plt
df.plot()
df[['ChestPainType' , 'RestingBP' , 'Cholesterol']]
df[['ChestPainType' , 'RestingBP' ]].plot()
df.plot(kind='scatter', x='ChestPainType' , y='RestingBP')
df.plot(kind='hist' , x='ChestPainType' , y='FastingBS')
df.plot(kind='scatter' , x='Age' , y='gender')
df.plot(kind='hist' , x='RestingECG' , y='MaxHR')
df.plot(kind='bar' , x='ExerciseAngina' , y='Oldpeak')
plt.show()
```
